```python
#get wholebrain data
#data = pd.read_csv('decoding_output/decoding_results_topics_wholebrain.csv')
data = pd.read_csv('decoding_output/decoding_results_topics_thresholded.csv')


#remove awkward brackets
#data['topic_num_name'] = data['topic_num_name'].str.replace("[","").str.strip()
#data['topic_num_name'] = data['topic_num_name'].str.replace("]","").str.strip()

#rank
data['scenario_rank'] = data['scenario'].rank(method='first', ascending=False)
data['evidence_rank'] = data['evidence'].rank(method='first', ascending=False)

#subset with topics from specific models
#utility
#002, #010, #086, #128, #186, #197
utility = data.loc[(data['topic_num'] == 2) | (data['topic_num'] == 10) | (data['topic_num'] == 86) |
                  (data['topic_num'] == 128) | (data['topic_num'] == 186) | (data['topic_num'] == 197)]
#affect
#126, #139, #180
affect = data.loc[(data['topic_num'] == 126) | (data['topic_num'] == 139) | (data['topic_num'] == 180)]

#moral judgement
#112, #135, #145, #154
moral_judgement = data.loc[(data['topic_num'] == 112) | (data['topic_num'] == 135) | (data['topic_num'] == 145) |
                  (data['topic_num'] == 154)]

#experience culture recall narrative
#112, #135, #145, #154
ecr_narrative = data.loc[(data['topic_num'] == 28) | (data['topic_num'] == 30) | (data['topic_num'] == 100) |
                  (data['topic_num'] == 111)]

#reading narrative
#009, #093, #095, #162
reading_narrative = data.loc[(data['topic_num'] == 9) | (data['topic_num'] == 93) | (data['topic_num'] == 95) |
                  (data['topic_num'] == 162)]

#wm narrative
#20, #179
wm_narrative = data.loc[(data['topic_num'] == 20) | (data['topic_num'] == 179)]
```
